```python
!pip install -q gradio ultralytics

import gradio as gr
from ultralytics import YOLO
from PIL import Image

# 1. تحميل الموديل
model = YOLO('yolov8n.pt')

# 2. تعريف وظيفة التوقع
def predict_image(img):
    results = model(img)
    # رسم المربعات على الصورة
    res_plotted = results[0].plot()
    return Image.fromarray(res_plotted)

# 3. بناء الواجهة
demo = gr.Interface(
    fn=predict_image,
    inputs=gr.Image(type="pil"),
    outputs=gr.Image(type="pil"),
    title="YOLOv8 Detection Anywhere",
    description="ارفعي صورتك هنا وهطلعلك الـ Detection في ثواني"
)

# 4. التشغيل مع تفعيل خاصية الـ Share
# السطر ده هو اللي بيطلع اللينك اللي آخره .gradio.live
demo.launch(share=True, debug=True)
```
